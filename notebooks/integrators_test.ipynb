サブモジュールのインポート

```python
try:
    from mylib import integrators as igt
    from mylib import models
except ImportError as e:
    print("Import failed:", e)
else:
    print("done")
```

グリッド生成

```python
#初期値グリッド
init_grid = {
    "x0": [10, 12],
    "y0": [2, 4],
}
init_names, init_values, init_sets = igt.make_init_sets(init_grid)

print(init_names)
print(init_values)
print(init_sets)

#パラメーターセットグリッド
param_grid = {
    "a": [0.3, 0.4, 0.5],
    "b": [0.1, 0.2, 0.3],
    "c": [0.3, 0.4, 0.5],
    "d": [1.3, 1.4, 1.5],
}
param_names, param_values, param_sets = igt.make_param_sets(param_grid)
print(param_names)
print(param_values)
print(param_sets)
```

数値計算

```python
dt = 0.001
t_max = 100

results = igt.run_param_loop(f=models.test_f,
                            g=models.test_g,
                            param_sets=param_sets,
                            init_sets=init_sets,
                            t_max=t_max,
                            dt=dt)
```

```python
#0番目と4番目のパラメータ,dictのキーの確認
print(results[0]["params"])
print(results[4]["params"])
print(results[0].keys())
```

```python
igt.save_experiment(
    results=results,
    param_grid=param_grid,
    init_grid=init_grid,
    model_name="test",
    dt=dt,
    t_max=t_max,
    exp_number=1,
)
```
